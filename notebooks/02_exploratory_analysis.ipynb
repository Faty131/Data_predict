```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/clean_data.csv")
```

```python
# Distribution des retards
plt.figure()
sns.histplot(df["delay_minutes"], bins=50)
plt.title("Distribution des retards")
plt.show()
```

```python
# Retards par heure
plt.figure()
sns.boxplot(x="hour", y="delay_minutes", data=df)
plt.title("Retards par heure")
plt.show()
```

```python
# Retards par statut
plt.figure()
sns.boxplot(x="Status", y="delay_minutes", data=df)
plt.xticks(rotation=45)
plt.title("Retards par statut")
plt.show()
```

```python
# Corrlation
plt.figure()
sns.heatmap(
    df[["delay_minutes","hour","dayofweek","month","heure_pointe"]].corr(),
    annot=True
)
plt.title("Corrlations")
plt.show()
```
